```python
INSTRUMENTS_DIR = '/ritter/share/data/IMAGEN/Instrument'
LABELS = ["Binge","Binge_growth",
          "AUDIT","AUDIT_quick","AUDIT_growth",
          "Combined_seo","Combined_ours",
          "Onset","Frequency","Amount"]
```

```python
def show_corr(instrument, session, ignore_cols=[], raw_vals=False, bin_val=False, vmax=0.4):
    df_all_labels = load_all_labels(raw_vals)
    if bin_val and not raw_vals:
        df_all_labels[LABELS] = df_all_labels[LABELS].applymap(lambda x: {0:0, 1:np.nan, 2:1}[x]).copy()
    instrument_df = pd.read_csv(f"{INSTRUMENTS_DIR}/{instrument}.csv").set_index('ID').groupby('Session').get_group(session)
    df = df_all_labels.join(instrument_df).copy()
    # remove the suffix 'mean' or 'sum' from the column names for better presentation
    df = df.rename(lambda c: c.replace(' mean', '').replace(' sum', '').replace('Sexsual', 'Sexual'), axis='columns')
    # drop columns like ['Sex', 'Site', 'Session'] before computing the correlation
    ignore_cols += ['Sex', 'Site', 'Session']
    df = df.drop(columns=ignore_cols)
    # compute the correlation and return the pandas Styler object
    return df.corr().apply(abs).style.format(
                    "{:.2f}").hide_columns( # remove traits from columns
                        [c for c in df.columns if (c not in LABELS) and (c not in ignore_cols)]).hide_index( # remove AAM labels from rows
                            [c for c in LABELS if c in df.columns]).background_gradient(
                                vmin=0, vmax=vmax)
```

## NEO

```python
show_corr(instrument='NEO', session='FU3',
          ignore_cols = ['Sex', 'Site'], # "Binge_growth","Combined_seo","Combined_ours", "AUDIT_growth", 'Onset'
          raw_vals = False, bin_val = True)
```
