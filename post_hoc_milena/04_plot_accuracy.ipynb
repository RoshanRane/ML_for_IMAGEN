```python
# import module
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
acc = pd.read_csv('/ritter/share/projects/Lab_Rotation_Milena/posthoc_analysis/accuracy.csv')
```

```python
acc_kSVM_FU3 = acc.loc[(acc['Model'] == 'Kernel SVM') & (acc['TP'] == 'FU3')]
```

```python
# make barplot and sort bars
sns.set(rc={'figure.figsize':(11.7,5)})
kSVM = sns.barplot(x='Confound',
            y="Accuracy Diff CB - BL", data=acc_kSVM_FU3,
            order=acc_kSVM_FU3.sort_values('Accuracy Diff CB - BL').Confound)
for item in kSVM.get_xticklabels():
    item.set_rotation(90)
```

```python
acc = pd.read_csv('/ritter/share/projects/Lab_Rotation_Milena/posthoc_analysis/accuracy.csv')
```

```python
acc_FU3 = acc.loc[(acc['Model'].isin(['Kernel SVM','Gradient Boosting'])) & (acc['TP'] == 'FU3')]
```

```python
# sort dataframe
acc_FU3 = acc_FU3.sort_values(['Model','Accuracy Diff CB - BL'])
```

```python
# make barplot and sort bars
title='Reductions in mean accuracy'
sns.set(rc={'figure.figsize':(11.7,4)})
sns.set_style("whitegrid")
FU3 = sns.barplot(x='Confound',
            y="Accuracy Diff CB - BL", hue='Model',
                   data=acc_FU3)
for item in FU3.get_xticklabels():
    item.set_rotation(90)
FU3.set_title(title)
```

```python
# sort dataframe
acc_FU3 = acc_FU3.sort_values(['Model','Accuracy CB'])
```

```python
# make barplot and sort bars
title='Mean accuracy after CB'
sns.set(rc={'figure.figsize':(11.7,4)})
sns.set_style("whitegrid")

FU3 = sns.lineplot(x='Confound',
            y="Accuracy CB", hue='Model',
                   data=acc_FU3, sort=True, marker = 'o')

for item in FU3.get_xticklabels():
    item.set_rotation(90)
    
FU3.set_title(title)
FU3.axhline(0.60,color='black',linewidth=1)
# plt.plot(['ESPAD_SubAbuse_Total','BIS & PSS & NEO Extro'], [0.678,0.678], linewidth=1.5)
# plt.plot(['ESPAD_SubAbuse_Total','BIS & PSS & NEO Extro'], [0.6929,0.6929], linewidth=1.5)
```
