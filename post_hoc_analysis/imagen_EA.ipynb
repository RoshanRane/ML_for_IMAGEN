# ERROR ANALYSIS
The preliminary results in our IMAGEN paper advocates for a more in-depth understanding of what contributes to the significant performance of the ML models for the three time-points. Error analysis of the model prediction to understand which and why the model performs well on some subjects and bad on some others with Instrument.

```python
from imagen_post_hoc_helper import *
```

```python
DATA_DIR = "/ritter/share/data/IMAGEN"
posthoc = IMAGEN_posthoc()
```

<b> Load the dataset </b>

```python
DF = posthoc.read_posthoc('IMAGEN_posthoc.csv')
```

```python
num = range(len(DF.columns))
col = list(DF.columns)
ROI = [list(x) for x in zip(num, col)]
# print(ROI)
```

<b> Generate the posthoc dataset </b>

```python
train = DF.groupby('Dataset').get_group('Training')
holdout  = DF.groupby('Dataset').get_group('Holdout')
Train_case = train.groupby('Model').get_group('GB')
Holdout = holdout.groupby('Model').get_group('GB')
Holdout_case = Holdout.groupby('Trial').get_group(0)
Case = pd.concat([Train_case, Holdout_case])
```

<b> Generate the columns </b>

```python
NEO = ['Openness mean', 'Conscientiousness mean', 'Extroversion mean',
       'Agreeableness mean', 'Neuroticism mean']
SURPS = ['Anxiety Sensitivity mean', 'Hopelessness mean',
         'Impulsivity mean', 'Sensation seeking mean']
CTQ = ['Emotional abuse sum', 'Physical abuse sum', 'Sexual abuse sum',
       'Emotional neglect sum', 'Physical neglect sum', 'Denial sum']
CTS = ['Assault mean', 'Injury mean', 'Negotiation mean',
       'Psychological Aggression mean', 'Sexual Coercion mean']
LEQ_v = ['Family valence','Accident valence', 'Sexuality valence',
         'Autonomy valence', 'Devience valence', 'Relocation valence',
         'Distress valence', 'Noscale valence', 'Overall valence']
LEQ_f = ['Family mean frequency', 'Accident mean frequency', 'Sexuality mean frequency',
         'Autonomy mean frequency', 'Devience mean frequency', 'Relocation mean frequency',
         'Distress mean frequency', 'Noscale mean frequency', 'Overall mean frequency']
BSI = ['Somatization mean', 'Obsession-Compulsion mean', 'Interpersonal Sensitivity mean',
       'Depression mean', 'Anxiety mean', 'Hostility mean', 'Phobic Anxiety mean',
       'Paranoid Ideation mean', 'Psychoticism mean', 'Positive Symptom Distress Index', 'Global Severity Index']
```
